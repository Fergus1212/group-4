```python
#Dependencies
import pandas_profiling as pp
from pandas_profiling import ProfileReport
import pandas as pd
from functools import reduce
```

```python
df = pd.read_csv("cleaned_data_09_19.csv")
```

```python
datetime = pd.to_datetime( df.loc[0:0]["Occurred Date"] )
```

```python
df["Datetime"] = pd.to_datetime( df["Occurred Date"])
```

```python
df = df.set_index("Datetime")
```

```python
df.to_csv("cleaned_data_with_datetime_09_19.csv")
```

```python
df.head()
```

```python
print("Incidents per year")
for year in range(2009,2019):
    print( f"Number of incidents in {year}: {len(df[df.index.year==year])}")
```

```python

for year in range(2009,2019):
    print( f'Family violence incidents in {year}: {len(df[ (df["Family Violence"]=="Y") & (df.index.year==year) ])}')
```

```python
for year in range(2009,2019):
    print( f'Number of clearances in {year}: {len(df[ (df["Clearance Status"]=="C") & (df.index.year==year) ])}')
```

```python
for year in range(2009,2019):
    print( f'Percentage of clearances in {year}: {len(df[ (df["Clearance Status"]=="C") & (df.index.year==year) ])/len(df[df.index.year==year])*100}')
```

```python
df[df.index.year==2009]["Highest Offense Description"].value_counts()[0:10]
```

```python
for year in range(2009,2019):
    print(f'Top ten offense types for {year}: \n\n{df[df.index.year==year]["Highest Offense Description"].value_counts()[0:10]}\n\n')
```

```python
for year in range(2009,2019):
    print(f'Top ten location types with most crimes reported in {year}: \n\n{df[df.index.year==year]["Location Type"].value_counts().iloc[0:10]}\n\n')
```

```python
for year in range(2009,2019):
    print(f'Top ten zip codes with most crimes reported in {year}: \n\n{df[df.index.year==year]["Zip Code"].value_counts().iloc[0:10]}\n\n')
```

```python
dow = ["Monday", "Tuesday", "Wednesday", "Thursday","Friday","Saturday","Sunday"]
print("Crimes reports by day of week")   
for day in range(0,7):
    print( f'Number of crime reports on {dow[day]}: {len(df[df.index.dayofweek==day])}')
```

```python
for year in range(2009,2019):
    df_for_year = df[df.index.year == year ]
    print(f"\nFor {year}\n")
    for day in range(0,7):
        print( f'Number of crime reports on {dow[day]}: {len(df_for_year[df_for_year.index.dayofweek==day])}')
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
print("Crimes reports by month of year")   
for month in range(1,13):
    print( f'Number of crime reports on {months[month-1]}: {len(df[df.index.month==month])}')
```
