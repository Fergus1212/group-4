### Analysis on Clearance Time by Area

```python
from pandas_profiling import ProfileReport
import pandas_profiling as pp
import pandas as pd
from functools import reduce
from datetime import date 
import matplotlib as plt
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
%matplotlib inline
```

```python
#Pulling in master dataframe
#drop 2019
crime_data = pd.read_csv("../Resources/clean_data_with_datetime_10_18.csv")
df = pd.DataFrame(crime_data)
#df.to_csv (r'Resources/cleaned_data.csv', index = False, header=True)
```

```python
#Convert columns to datetime
df['Occurred Date'] =pd.to_datetime(df['Occurred Date'])
df['Clearance Date'] =pd.to_datetime(df['Clearance Date'])
```

```python
#creating function to find differnce between two dates
def numOfDays(date1, date2): 
    return (date2-date1)
```

```python
#creating function to find differnce between two dates
df['Days to Clear Case']=numOfDays(df['Occurred Date'],df['Clearance Date'])
df['Days to Clear Case']=df['Days to Clear Case'].dt.days
```

```python
#create group by for sectorsse
sector_grp = df.groupby(['APD Sector']).mean()
objects =sector_grp.index.tolist()
index = list(sector_grp.index) 
```

```python
#Bar graph comparing each sector days to clear case
ypos = np.arange(len(objects))
plt.bar(ypos,sector_grp['Days to Clear Case'])
plt.xticks(ypos, index)
plt.ylabel('Average Days to Clear a Case')
plt.title('Average Days to Clear Case by Sector')
plt.show();
```

```python
# through research we discover that 88 means outside of austin. IDA, UT, Henry, and 88 are removed from dataset
df=df[df['APD Sector'] != '88']
df=df[df['APD Sector'] != 'HENRY']
df=df[df['APD Sector'] != 'IDA']
df=df[df['APD Sector'] != 'UT']
```

```python
#create bar graph again. this time sans IDA, UT, Henry, and 88
sector_grp = df.groupby(['APD Sector']).mean()
objects =sector_grp.index.tolist()
index = list(sector_grp.index) 
ypos = np.arange(len(objects))
plt.bar(ypos,sector_grp['Days to Clear Case'])
plt.xticks(ypos, index)
plt.ylabel('Average Days to Clear a Case')
plt.title('Average Days to Clear Case by Sector')
plt.show();
```

```python
sector_grp = df.groupby(['APD Sector']).mean()
objects =sector_grp.index.tolist()
index = list(sector_grp.index) 
ypos = np.arange(len(objects))
sector_bar =plt.bar(ypos,sector_grp['Days to Clear Case'])
plt.xticks(ypos, index)
plt.ylabel('Average Days to Clear a Case')
plt.title('Average Days to Clear Case by Sector')
sector_bar[7].set_color('r')
plt.show();
```

```python
sector_grp = df.groupby(['APD Sector']).count()
objects =sector_grp.index.tolist()
index = list(sector_grp.index) 
ypos = np.arange(len(objects))
sector_bar =plt.bar(ypos,sector_grp['Days to Clear Case'])
plt.xticks(ypos, index)
plt.ylabel('Case Count')
plt.title('Count of Cases by Sector')
sector_bar[7].set_color('r')
plt.show();
```

```python
#Average number of cases in each sector were actually cleared? 

df['Clearance flag']=df['Clearance Status'].replace(np.nan,"Not Cleared")
df['Clearance flag']=df['Clearance flag'].replace('C',"Cleared")
df['Clearance flag']=df['Clearance flag'].replace('N',"Cleared")
df['Clearance flag']=df['Clearance flag'].replace('O',"Cleared")
df.head()
```

```python
clear_grp = df.groupby(['Clearance flag','APD Sector']).count()
clear_grp.to_csv('')
```

```python
c_df = clear_grp.drop(['Not Cleared'])
c_df.sort_values(by=['APD Sector'])
n_df= clear_grp.drop(['Cleared'])
n_df.sort_values(by=['APD Sector'])
c_df.to_csv('Cleared Count.csv')
n_df.to_csv('Not Cleared Count.csv')
```

```python
c_df
```

```python
#Delta between cleared and not clreaed for each sector over 8 years
combo_df = c_df.merge(n_df, left_on ='APD Sector',right_on ='APD Sector',suffixes=('_left', '_right'))
combo_df['Delta'] = combo_df['Datetime_left']-combo_df['Zip Code_right']
combo_df['Total Cases'] = combo_df['Datetime_left']+combo_df['Zip Code_right']
combo_df['Clear Percentage'] = (combo_df['Datetime_left']/combo_df['Total Cases'])*100
combo_df
```

```python
objects =c_df.index.tolist()
width = 0.25
index = list(n_df.index)  
index =pd.DataFrame(index)
ypos = np.arange(len(objects))
plt.bar(ypos,c_df['Datetime'],width,label ='Cleared')

plt.ylabel('Count of Cases')
plt.title('Cleared vs Not Cleared by Sector')

plt.bar([p + width for p in ypos],n_df ['Datetime'],width)
plt.xticks(ypos, index[1], rotation=45)
#ax = fig.add_subplot(111) # Create matplotlib axes
#ax2 = ax.twinx()

plt.legend(['Cleared', 'Not Cleared'], loc='upper left')
```
