```python

def drawHeatMap(location, val, zoom, radius):
    # setting the data and parameters
    heatmap_layer = gmaps.heatmap_layer(locations, val, dissipating = False)
        #heatmap_layer = gmaps.WeightedHeatmap(data=val)
    heatmap_layer.point_radius = radius
    # draw the heatmap into a figure
    fig = gmaps.figure()
    fig = gmaps.figure(center = [center_lat,center_lng], zoom_level=zoom)
    fig.add_layer(heatmap_layer)
    return fig

```

```python
#Average number of cases in each sector were actually cleared? 

df['Clearance flag']=df['Clearance Status'].replace(np.nan,"Not Cleared")
df['Clearance flag']=df['Clearance flag'].replace('C',"Cleared")
df['Clearance flag']=df['Clearance flag'].replace('N',"Cleared")
df['Clearance flag']=df['Clearance flag'].replace('O',"Cleared")
df.head()
```

```python
clear_grp = df.groupby(['Clearance flag','APD Sector']).count()
clear_grp.to_csv('')
```

```python
c_df = clear_grp.drop(['Not Cleared'])
c_df.sort_values(by=['APD Sector'])
n_df= clear_grp.drop(['Cleared'])
n_df.sort_values(by=['APD Sector'])
c_df.to_csv('Cleared Count.csv')
n_df.to_csv('Not Cleared Count.csv')
```

```python
c_df
```

```python
#Delta between cleared and not clreaed for each sector over 8 years
combo_df = c_df.merge(n_df, left_on ='APD Sector',right_on ='APD Sector',suffixes=('_left', '_right'))
combo_df['Delta'] = combo_df['Datetime_left']-combo_df['Zip Code_right']
combo_df['Total Cases'] = combo_df['Datetime_left']+combo_df['Zip Code_right']
combo_df['Clear Percentage'] = (combo_df['Datetime_left']/combo_df['Total Cases'])*100
combo_df
```

```python
objects =c_df.index.tolist()
width = 0.25
index = list(n_df.index)  
index =pd.DataFrame(index)
ypos = np.arange(len(objects))
plt.bar(ypos,c_df['Datetime'],width,label ='Cleared')

plt.ylabel('Count of Cases')
plt.title('Cleared vs Not Cleared by Sector')

plt.bar([p + width for p in ypos],n_df ['Datetime'],width)
plt.xticks(ypos, index[1], rotation=45)
#ax = fig.add_subplot(111) # Create matplotlib axes
#ax2 = ax.twinx()

plt.legend(['Cleared', 'Not Cleared'], loc='upper left')
```
